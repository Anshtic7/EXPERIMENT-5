```python
# Creation of Dictionary
dihdict = {
    "brand" : "Ford", "model" : "Mustang", "year" : 1964
}
print(dihdict)
```

```python
# Retrieving values from dictionary
dihdict = {
    "brand" : "Ford", "model" : "Mustang", "year" : 1964
}
print(dihdict["brand"])
```

```python
# Dictionary takes latest value given for key. It will not allow duplicates.
dihdict = {
    "brand" : "Ford", "model" : "Mustang", "year" : 1964, "year" : 2020
}
print(dihdict)
# length of key-value pairs
len(dihdict)
```

```python
# Changing/Adding key:value in dictionary
car = {
    "brand" : "Ford", "model" : "Mustang", "year" : 1964
}
car["color"] = "red"
print(car)
x = car.values()
print(x) #after the change
print(type(car))
```

```python
# Retrival of values
student = {
    "roll" : 101, "name" : "Ansh", "branch" : "ENTC"
}
student.pop("branch")
print("After removal:", student)
```

'''PROBLEM STATEMENT 3: A DICTIONARY STORES USERNAMES AND PASSWORDS.

WRITE A PYTHON PROGRAM TO VALIDATE USER LOGIN.'''

```python
# Sample dictionary of usernames and passwords
user_credentials = {
    "user1": "pass123",
    "admin": "secureadmin",
    "guest": "guestpass"
}

# Get username and password from the user
username = input("Enter your username: ")
password = input("Enter your password: ")

# Validate login using if else statements
if user_credentials.get(username) == password:
  print("Login successful!")
else:
  print("Invalid username or password.")
```

'''PROBLEM STATEMENT 4: A DICTIONARY STORES STUDENT NAMES AND MARKS.

WRITE A PYTHON PROGRAM TO FIND THE STUDENT WHO SCORED THE HIGHEST MARKS.'''

```python
marks = {"A" : 85, "B" : 95, "C" : 70, "D" : 65, "E" : 90}
topper = max(marks, key = marks.get)
print("Topper:", topper)
print("Marks:" , marks[topper])
```
